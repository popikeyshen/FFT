FFT samples

sin, step function, dirac func, stop sin


```python
import matplotlib.pyplot as plt
import numpy as np

f   = 20
f_s = 200

t = np.linspace(0, 2, 2 * f_s, endpoint=False)

sin = np.sin(f * 2 * np.pi * t)

step = np.linspace(0, 2, 2 * f_s, endpoint=False)
for i in range(0,len(t)/2):
        step[i]=1
for i in range(len(t)/2,len(t)):
        step[i]=0

stop = np.sin(f * 2 * np.pi * t)
for i in range(len(stop)/2,len(stop)):
        stop[i]=0
            

fig, ax = plt.subplots()
ax.plot(t, sin)

fig, ax = plt.subplots()
ax.plot(t, step)

fig, ax = plt.subplots()
ax.plot(t, stop)

plt.show()
```

Make fft

```python
from scipy import fftpack

S = fftpack.fft(sin)
freqs = fftpack.fftfreq(len(sin)) * f_s

fig, ax = plt.subplots()
ax.stem(freqs, np.abs(S))



ST = fftpack.fft(step)
freqs = fftpack.fftfreq(len(step)) * f_s

fig, ax = plt.subplots()
ax.stem(freqs, np.abs(ST))



X = fftpack.fft(stop)
freqs = fftpack.fftfreq(len(stop)) * f_s

fig, ax = plt.subplots()
ax.stem(freqs, np.abs(X))

XX=X.copy()
for i in range(len(X)/2,len(X)):
    XX[i]=0

fig, ax = plt.subplots()
ax.stem(freqs, np.abs(XX))

plt.show()
```

spectrum to signal

```python


my_fft=[0,2,0,0,0,0,0,0,0,0,0]
irfft_output = np.fft.irfft(my_fft)

fig, ax = plt.subplots()
ax.stem(irfft_output)


irfft_output = np.fft.ifft(my_fft)

fig, ax = plt.subplots()
ax.stem(irfft_output)



irfft_output = np.fft.irfft(X)

fig, ax = plt.subplots()
ax.plot(irfft_output[:200])


irfft_output = np.fft.irfft(XX)

fig, ax = plt.subplots()
ax.plot(irfft_output[:200])


plt.show()
```
